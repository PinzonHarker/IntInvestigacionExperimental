# Acerca del proyecto
Se tendran algunas medidas y graficas para el análisis del problema de la difusión.

```python
%matplotlib widget
import numpy as np
import matplotlib.pyplot as plt
# style
import scienceplots
plt.style.use(['science', 'notebook'])
plt.rcParams['text.usetex'] = True

def n(x, t, A0, D):
    return A0 / (2 * np.sqrt(np.pi * D * t)) * np.exp(-x**2 / (4 * D * t))

fig, ax = plt.subplots()
t = np.array([0.1, 0.25, 0.5, 0.75, 1, 1.5])
x = np.linspace(-4, 4, 1000)
for times in t:
    y = n(x, times, 1, 1)
    ax.plot(x, y, label=fr'$\tau$ = {times}')
ax.axvline(x=0, color='black', linestyle='-', linewidth=1)
ax.axhline(y=0, color='black', lw=1)
ax.set_xlabel('x')
ax.set_ylabel('n(x, t)')
ax.set_xlim(-4, 4)
ax.set_ylim(bottom=0)
ax.legend()
#fig.savefig('n(x,t)_vs_x.pdf')
plt.show()
```

```python
# Difusión bidimensional
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
# style
import scienceplots
plt.style.use(['science', 'notebook', 'high-vis'])
plt.rcParams['text.usetex'] = True

def n(r, t, A0, D):
    return A0 / (2 * D * t) * np.exp(-r**2 / (4 * D * t))

fig, ax = plt.subplots(figsize=(8,4.5))
t = np.array([0.1, 0.25, 0.5, 0.75, 1, 1.5])
x = np.linspace(-4, 4, 1000)
for times in t:
    y = n(x, times, 1, 1)
    ax.plot(x, y, label=fr'$\tau$ = {times}')
ax.axvline(x=0, color='black', linestyle='-', linewidth=1)
ax.axhline(y=0, color='black', lw=1)
ax.set_xlabel(r'$r$')
ax.set_ylabel(r'n($r$, t)')
ax.set_xlim(0, 2)
ax.set_ylim(bottom=0)
ax.legend()
fig.savefig('n(r,t)_vs_r.pdf')
plt.show()
```
